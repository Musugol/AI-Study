```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
california.keys()
```

```python
print(california.data.shape, california.target.shape)
print(california.feature_names)
```

```python
dfx = pd.DataFrame(california.data, columns=california.feature_names)
dfy = pd.DataFrame(california.target, columns=["MEDV"])
california_df = pd.concat([dfx, dfy], axis=1)
california_df.tail()
```

```python
from sklearn.model_selection import train_test_split
x = california_df[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']]
y = california_df[['MEDV']]
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
mtrg = LinearRegression()
mtrg.fit(x_train, y_train)
```

```python
my_apt = [[1.87, 17, 6, 2, 950, 2.45, 39.43, -121.19]]
my_predict = mtrg.predict(my_apt)
print(my_predict)
```
